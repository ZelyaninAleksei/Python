```python
import seaborn as sns
import pandas as pd
penguins = sns.load_dataset("penguins")
penguins.head(10)
```

- species - информация о виде пингвина;
- island - название острова в архипелаге Палмера;
- bill_length_mm - длина клюва в мм;
- bill_depth_mm - толщина клюва в мм;
- flipper_length_mm - длина крыла в мм;
- body_mass_g - масса тела в граммах;
- sex - пол особи.

```python
sns.scatterplot(data=penguins, x='body_mass_g', y='flipper_length_mm')
```

```python
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', size='body_mass_g')
```

```python
lst = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'sex']
g = sns.PairGrid(penguins[lst])
g.map(sns.scatterplot)
```

```python
sns.heatmap(penguins.corr(), annot = True)
```

```python
sns.histplot(data=penguins, x='bill_depth_mm')
```

```python
sns.histplot(data=penguins, x='species')
```
